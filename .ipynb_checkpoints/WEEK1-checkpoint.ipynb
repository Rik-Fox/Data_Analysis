```julia
n = 10
a = 1.5

x = range(0,stop=n)
dens = zeros(length(x))
CDF = zeros(length(x))

for i=1:length(x)
    
    dens[i] = (exp(-(x[i]/a)))/a
    CDF[i] = (1-exp(-(x[i]/a)))
end

#figure()
plot(x,dens, label = "Exp Density",size=(800,500))
plot!(x,CDF, label = "Exp CDF")
title!("Culmative Density Function and Density Function of Exponential Distribution")
xlabel!("X")
```
